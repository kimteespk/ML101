```python
import pandas
import yfinance as yf
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = yf.download(
    'BBL.BK KBANK.BK SCB.BK BAY.BK TTB.BK KKP.BK TISCO.BK SAWAD.BK MTC.BK TIDLOR.BK AMANAH.BK ASK.BK CHAYO.BK JMT.BK KTC.BK JMT.BK MICRO.BK SAK.BK THANI.BK',
    start = '2018-01-01', end = '2021-10-14', interval= '1d')
```

```python
df = data['Adj Close'].copy()
```

```python
# too much null cant not .notnull only one column
data.loc[df['CHAYO.BK'].notnull()]
```

```python
# find the first data row that not null
df.loc[~df.isnull().sum(1).astype(bool)].iloc[0]
```

```python
df = df.fillna(method= 'ffill')
df
```

```python
df_ret =df.pct_change()
```

```python
df_ret = df_ret.loc[~df_ret.isnull().sum(1).astype(bool)]
```

```python
df_ret.corr()
```

```python
g = sns.clustermap(df_ret.corr(), method='ward', center = 0, cmap="RdYlGn",
                   dendrogram_ratio=(.1, .1),
                   linewidths=.75, figsize=(12, 13),
                   annot = True)
```

```python
group1 = df[['SCB.BK', 'TTB.BK', 'BBL.BK', 'KBANK.BK', 'BAY.BK', 'TISCO.BK']]
group2 = df[['AMANAH.BK', 'TIDLOR.BK', 'MTC.BK', 'SAWAD.BK', 'KTC.BK', 'JMT.BK', 'THANI.BK']]
```

```python
group1_geo_ret = (group1.pct_change()+1).cumprod()
group2_geo_ret = (group2.pct_change()+1).cumprod()
```

```python
group1_geo_ret.plot(figsize=(10,6))
group2_geo_ret.plot(figsize=(10,6))
plt.show()
```
