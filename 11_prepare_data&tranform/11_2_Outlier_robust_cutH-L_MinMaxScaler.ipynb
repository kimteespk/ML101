# Min-Max Scaler

```python
df_x_normalized = pd.DataFrame(X, columns= data.data.columns)
```

```python
df_x_normalized.describe()
```

```python
upper = df_x_normalized.quantile(0.75)
upper
# just check that pencentile is correct
```

## Pandas Style

```python
# set high and low pencentile
low = 0.05
high =  .95
quant_df = df_x_normalized.quantile([low,high])
quant_df
# latitude and longtitude are droped        
```

```python
# use lambda function to get data in the quantiled  zone
filt_df_x = df_x_normalized.apply(
    lambda x: x[(x >= quant_df.loc[low,x.name]) & # loc row from lower by using a QUANT_DF as จุดอ้างอิงสำหรับขอบบนล่าง
    (x <= quant_df.loc[high, x.name])], axis= 0 
)
```

```python
filt_df_x
```

```python
filt_df_x.dropna()
```

## Min_Max programming metrix style

```python
Q1 = df_x_normalized.quantile(0.05)
Q3 = df_x_normalized.quantile(0.95)
```

```python
filt_df_x2 = df_x_normalized[~
    ((df_x_normalized < Q1) |               # get data from Q1 to Q3 by use ~ to invesre operation
    (df_x_normalized > Q3 )).any(axis= 1)   # axis= 1 is mean look into the row
]

```

```python
(df_x_normalized < Q1) | (df_x_normalized > Q3) # if data in a row <q1 or >Q3 = True
# | = or
```

```python
# WRONG AXIS EXSAMPLE
~((df_x_normalized < Q1) | (df_x_normalized > Q3)).any()
# any(axis= 0) its gonna be summed by column view 
# axis = is defualt arg
```

```python
# inverse operating
~((df_x_normalized < Q1) | (df_x_normalized > Q3)).any(axis =1) 
```

```python
df_x_normalized[~((df_x_normalized < Q1) | (df_x_normalized > Q3)).any(axis =1)]
# loc main df data in the zone limited by Q1 or Q3
# **** dont use >= or <= because this use inverse operating funcion
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler =  MinMaxScaler()
scaler.fit(df_x_normalized)
df_x_scaled = scaler.transform(df_x_normalized)
```
