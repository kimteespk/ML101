```python
from sklearn.preprocessing import StandardScaler
from sklearn.datasets import fetch_california_housing
import pandas as pd
```

```python
data = fetch_california_housing(as_frame=True)
```

```python
X = data.data
y = data.target
```

```python
df_x = pd.DataFrame(X, columns= data.data.columns)
```

```python
scaler = StandardScaler()
scaler.fit(df_x)
df_x_std = scaler.transform(df_x)
```

```python
# change format from np.array to Dataframe
df_x_std = pd.DataFrame(df_x_std, columns= X.columns)
df_x_std
```

```python
df_x2 = df_x.copy()
df_x2.values[:] = scaler.fit_transform(df_x2)
df_x2
```
